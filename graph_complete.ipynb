```python
import matplotlib.pyplot as plt

meses = ['Jan', 'Fev', 'Mar', 'Abr', 'Mai', 'Jun', 'Jul', 'Ago', 'Set', 'Out', 'Nov', 'Dez']
vendas = [85, 90, 78, 92, 88, 96, 94, 98, 100, 102, 105, 108]

fig, ax = plt.subplots()

ax.plot(meses, vendas)
ax.set_xlabel('Meses')
ax.set_ylabel('Vendas')
ax.set_title('Evolução de vendas ao longo do ano.')


plt.show()
```

```python
# Data
produtos = ['Produto A', 'Produto B', 'Produto C', 'Produto D', 'Produto E']
quantidades = [23, 45, 17, 30, 37]

fix, ax = plt.subplots()

ax.bar(produtos, quantidades)
ax.set_xlabel('Produtos')
ax.set_ylabel('Quantidades')
ax.set_title('Quantidade de vendas por produtos')


plt.show()
```

```python
# Data

categorias = ['Eletronicos', 'Roupas', 'Alimentos', 'Livros']
quantidades_pizza = [45, 30, 15, 10]
destaque = [0.1, 0.05, 0.05, 0.05]

fig, ax = plt.subplots()


ax.pie(quantidades_pizza, labels=categorias, explode=destaque, autopct='%.2f')
ax.set_title('Categorias de produtos vendidos')


plt.show()
```

```python
# Data

anos = [2018, 2019, 2020, 2021]
vendas_online = [5, 10, 15, 20]
vendas_loja = [10, 8, 6, 4]

fig, ax = plt.subplots()

ax.bar(anos, vendas_online, label='Online')
ax.bar(anos, vendas_loja, bottom=vendas_online, label='Loja')

ax.set_xticks(anos)

ax.legend()

ax.set_xlabel('Ano')
ax.set_ylabel('Vendas em milhões de Reais')
ax.set_title('Vendas - Online X Loja')



plt.show()
```
